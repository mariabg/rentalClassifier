```python
fHost_is_superhost = tf.feature_column.numeric_column("host_is_superhost")
fSecurity_deposit = tf.feature_column.numeric_column("security_deposit")
fSeaViews = tf.feature_column.numeric_column("seaViews")
fBathrooms = tf.feature_column.numeric_column("bathrooms")
fBedrooms = tf.feature_column.numeric_column("bedrooms")
fBeds = tf.feature_column.numeric_column("beds")
fGuests_included = tf.feature_column.numeric_column("guests_included")
fReview_scores_rating = tf.feature_column.numeric_column("review_scores_rating")
fAvailability = tf.feature_column.numeric_column("availability")

features = [fHost_is_superhost, fSecurity_deposit, fSeaViews, fBathrooms, fBedrooms, fBeds, fGuests_included, fReview_scores_rating, fAvailability]
```

```python
classifier = tf.estimator.LinearClassifier(
    feature_columns=features,
    n_classes=3,
    model_dir="/tmp/predictPriceGroup")
```

```python
def input_fn(dataset):
    def _fn():
        features = {
            "host_is_superhost": tf.constant(dataset["host_is_superhost"]),
            "security_deposit": tf.constant(dataset["security_deposit"]),
            "seaViews": tf.constant(dataset["seaViews"]),
            "bathrooms": tf.constant(dataset["bathrooms"]),
            "bedrooms": tf.constant(dataset["bedrooms"]),
            "beds": tf.constant(dataset["beds"]),
            "guests_included": tf.constant(dataset["guests_included"]),
            "review_scores_rating": tf.constant(dataset["review_scores_rating"]),
            "availability": tf.constant(dataset["availability"])
        }
        label = tf.constant(groupedPriceArray)
        return features, label
    return _fn
```

```python
classifier.train(input_fn=input_fn(dataset), steps=10000)
print('fit done')
```

```python
accuracy_score = classifier.evaluate(input_fn=input_fn(test), 
                                     steps=100)["accuracy"]
print('\nAccuracy: {0:f}'.format(accuracy_score))
```
