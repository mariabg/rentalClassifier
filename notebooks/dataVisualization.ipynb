```python
import sys
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# df = pd.read_csv('../data/features.csv');
df = pd.read_csv('../data/norm_all_data.csv');
```

### Overall visualizations

```python
sns.distplot(df['normPrice']);
```

```python
sns.distplot(df['availability'].dropna());
```

```python
sns.set()
df['groupedPrice'] = df['groupedPrice'].astype(str).apply(lambda x: (x.replace('lowCost', '0'))).astype(str).apply(lambda x: (x.replace('medium', '1'))).astype(str).apply(lambda x: (x.replace('premium', '2')))
cols = ['normPrice', 'beds', 'bedrooms', 'bathrooms', 'property_type', 'availability', 'groupedPrice', 'seaViews']
sns.pairplot(df[cols].dropna(), size = 2.5)
plt.savefig('../outcomes/general_vis.png')
```
